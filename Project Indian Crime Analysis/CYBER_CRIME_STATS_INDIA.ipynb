<h2><a id='division_ID3'>Data Visualization</a></h2>

[MENU](#division_ID0)

- <h3><b>2013 Cyber Crime Statewise Comparison</b></h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(16, 9))
plt.title('2013 Cyber Crime Statewise Comparison', bbox={'facecolor': '0.8', 'pad': 10}, loc='center')
sns.barplot(x=df.iloc[:, 0], y=df.iloc[:, -1], hue=df.iloc[:, 0], palette='rocket', legend=False)
plt.ylabel('No. of Cyber Crimes')
plt.xticks(rotation=90)
plt.show()
```

- <h3><b>2013 Cyber Crime [Fraud / Illegal Gain] Statewise Comparison</b></h3>

```python
import random
plt.figure(figsize = (9, 9))
colors = [(random.random(), random.random(), random.random()) for i in range(25)]
plt.title('2013 Cyber Crime [Motive: Fraud / Illegel Gain] Statewise', bbox={'facecolor':'0.8', 'pad':10}, loc = 'center')
plt.pie(list(df.iloc[:, -4]), colors = colors)
my_circle=plt.Circle((0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(list(df.iloc[:, 0]), loc = 'upper right', bbox_to_anchor = (-0.1, 1.), fontsize=15, ncol = 2)
plt.show()
```

- <h3><b>2013 Cyber Crime [Eve Teasing] Statewise Comparison</b></h3>

```python
plt.figure(figsize = (9, 9))
colors = [(random.random(), random.random(), random.random()) for i in range(25)]
plt.title('2013 Cyber Crime [Motive: Eve Teasing] Statewise', bbox={'facecolor':'0.8', 'pad':10}, loc = 'center')
plt.pie(list(df.iloc[:, -3]), colors = colors)
my_circle=plt.Circle((0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(list(df.iloc[:, 0]), loc = 'upper right', bbox_to_anchor = (-0.1, 1.), fontsize=15, ncol = 2)
plt.show()
```

- <h3><b>Statewise Instensity of Cyber Crime 2013</b></h3>

```python
import geopandas as gpd
df = pd.DataFrame(list(zip(list(df.iloc[:, 0]), list(df.iloc[:, -1]))), columns = ['Area_Name', 'Cyber Crime'])
map_file = gpd.read_file(r"E:\project-gtp-main\.mapDependencies\Indian_States.shp")
merged = map_file.set_index('st_nm').join(df.set_index('Area_Name'))
na = ['Andaman & Nicobar Island', 'Arunanchal Pradesh', 'Dadara & Nagar Havelli', 'Daman & Diu', 
      'NCT of Delhi', 'Jammu & Kashmir', 'Telangana', 'Odisha']
val = [18, 10, 0, 1, 150, 46, 0, 104]
j = 0
for i in na:
    merged.at[i, list(merged.columns)[1]] = val[j]
    j += 1
fig, ax = plt.subplots(1, figsize=(16, 16))
ax.axis('off')
ax.set_title('2013 Statewise Cyber Crime', bbox={'facecolor':'0.8', 'pad':5}, loc = 'center')
fig = merged.plot(column='Cyber Crime', cmap='Spectral', linewidth=0.5, ax=ax, edgecolor='0.2',legend=True)
```
