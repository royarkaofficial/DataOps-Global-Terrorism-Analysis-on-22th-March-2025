<h2><a id='division_ID2'>Data Preprocessing</a></h2>

[MENU](#division_ID0)

```python
# check for null values in the dataframe
df.isnull().sum()
```

```python
# fill the null values with 0
df.fillna(0, inplace = True)
```

```python
# drop unneccesary column
df.drop(['Group_Name', 'Sub_Group_Name'], axis = 1, inplace = True)
```

```python
# change column datatypes to integer
df.iloc[:, [2, 4, 5, 6, 7, 8]] = df.iloc[:, [2, 4, 5, 6, 7, 8]].astype(int)
```

```python
# rename columns
dict_col = {
    'Victims_Above_50_Yrs': '>50 years',
    'Victims_Upto_10_15_Yrs': '10-15 years',
    'Victims_Upto_10_Yrs': '<10 years',
    'Victims_Upto_15_18_Yrs': '15-18 years',
    'Victims_Upto_18_30_Yrs': '18-30 years',
    'Victims_Upto_30_50_Yrs': '30-50 years'
}
murder = df.rename(columns = dict_col)
murder
```

<h2><a id='division_ID3'>Data Visualization</a></h2>

[MENU](#division_ID0)

- <h3><b>2001 - 2010 Age classes of Murder Victims</b></h3>

```python
import matplotlib.pyplot as plt
indices = [5, 4, 6, 7, 8, 2]
victims = [murder.iloc[:, i].sum() for i in indices]
victim_ages = [murder.columns[i] for i in indices]
colors = ['blue', 'teal', 'yellow', 'crimson', 'maroon', 'darkolivegreen']
plt.figure(figsize=(16, 9))
plt.title('Murder Victims Age Distribution', bbox={'facecolor': '0.8', 'pad': 10}, loc='center')
plt.pie(victims, colors=colors, wedgeprops={'edgecolor': 'white'})
plt.gca().add_artist(plt.Circle((0, 0), 0.7, color='white'))  # Creating a donut chart
plt.legend(victim_ages, loc='center right', bbox_to_anchor=(-0.1, 1.), fontsize=20)
plt.show()
```

- <h3><b>Statewise comparison of 2001 - 2010 Murders</b></h3>

```python
import seaborn as sns
area_names = df.iloc[:, 0].unique()
murders = [df[df.iloc[:, 0] == area].iloc[:, 3].sum() for area in area_names]
sns.set_style("dark")
plt.figure(figsize=(16, 9))
sns.barplot(y=area_names, x=murders, hue=area_names, palette='flare', legend=False)
plt.xlabel('No. of Murders')
plt.title('Statewise 2001 - 2010 Murders', bbox={'facecolor': '0.8', 'pad': 10}, loc='center')
plt.show()
```

- <h3><b>2001 - 2010 yearwise Murder (Statewise)</b></h3>

```python
def state_murder_stats(state_name):
    new = df[df.iloc[:, 0] == state_name]
    timeline = df.iloc[:, 1].unique()
    yearwise_crime = [new[new.iloc[:, 1] == year].iloc[:, 3].sum() for year in timeline]
    plt.figure(figsize=(16, 9))
    sns.barplot(x=timeline, y=yearwise_crime, hue=timeline, palette='RdYlGn_r', legend=False)
    plt.xlabel('Timeline')
    plt.ylabel('No. of Murders')
    plt.title(f'{state_name} 2001 - 2010 Murder Statistics', bbox={'facecolor': '0.8', 'pad': 10}, loc='center')
    plt.show()
for i in area_names:
    state_murder_stats(i)
```

- <h3><b>Statewise Comparison of Yearwise Murder 2001 - 2010</b></h3>

```python
def plot_detail(year):
    states = murder.iloc[:, 0].unique()
    murders = [murder[(murder.iloc[:, 0] == state) & (murder.iloc[:, 1] == year)]['Victims_Total'].sum() for state in states]
    plt.figure(figsize=(16, 9))
    sns.barplot(x=states, y=murders, palette='rocket')
    plt.xticks(rotation=90)
    plt.ylabel('No. of Murders')
    plt.title(f'{year} Statewise Murder Comparison', bbox={'facecolor': '0.8', 'pad': 10}, loc='center')
    plt.show()
for i in range(2001, 2011):
    plot_detail(i)
```
