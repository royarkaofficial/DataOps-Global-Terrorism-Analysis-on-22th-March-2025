<h2><a id='division_ID1'>Data Collection</a></h2>

[MENU](#division_ID0)

```python
# import datasets from the local repository
import pandas as pd
df1 = pd.read_csv(r"E:\project-gtp-main\Dataset\2016 Cases against Police Personnels.csv")
df2 = pd.read_csv(r"E:\project-gtp-main\Dataset\2017 Cases against Police Personnels.csv")
df3 = pd.read_csv(r"E:\project-gtp-main\Dataset\2018 Cases against Police Personnels.csv")
```

<h2><a id='division_ID2'>Data Preprocessing</a></h2>

[MENU](#division_ID0)

```python
# check null values in the dataset
print(df1.isnull().sum())
print(df2.isnull().sum())
print(df3.isnull().sum())
```

```python
# preprocess df1
df1 = df1.iloc[:, [2, 3, 4, 5, 6]]
df1.rename(columns={'Total Number of Criminal Cases Registered during the year': 'Total Cases', 'Police Personnel Arrested': 'Police Arrested', 'Cases Chargesheeted': 'Chargesheets', 'Police Personnel Chargesheeted': 'Chargesheet against police'}, inplace = True)
df1
```

```python
# preprocess df2
df2 = df2.iloc[:, [2, 3, 7, 5, 8]]
df2.rename(columns={'Number of Cases - Registered': 'Total Cases', 'Number of Police Personel - Arrested': 'Police Arrested', 'Number of Cases - Chargesheeted': 'Chargesheets', 'Number of Police Personel - Chargesheeted': 'Chargesheet against police'}, inplace = True)
df2
```

```python
# preprocess df3
df3 = df3.iloc[:, [2, 3, 7, 5, 8]]
df3.rename(columns={'Number of Cases - Registered': 'Total Cases', 'Number of Police Personel - Arrested': 'Police Arrested', 'Number of Cases - Chargesheeted': 'Chargesheets', 'Number of Police Personel - Chargesheeted': 'Chargesheet against police'}, inplace = True)
df3
```

<h2><a id='division_ID3'>Data Visualization</a></h2>

[MENU](#division_ID0)

- <h3><b>2016 - 2018 Chargesheets Against Police Statewise Overall Comaparison</b></h3>

```python
import matplotlib.pyplot as plt
import numpy as np
areas = df1.iloc[:, 0].tolist()
r = np.arange(len(areas))
w = 0.3
plt.figure(figsize=(12, 6))
for i, df in enumerate([df1, df2, df3]):
    plt.bar(r + i * w, df.iloc[:, -1], w)
plt.xticks(r + w, areas, rotation=90)
plt.title('2016 - 2018 Chargesheets Against Police', bbox={'facecolor': '0.8', 'pad': 5})
plt.legend(['2016', '2017', '2018'])
plt.show()
```

- <h3><b>2016 - 2018 Cases Against Police Statewise Overall Comaparison</b></h3>

```python
r = np.arange(len(areas))
w = 0.3
plt.figure(figsize=(12, 6))
for i, df in enumerate([df1, df2, df3]):
    plt.bar(r + i * w, df.iloc[:, 1], w)
plt.xticks(r + w, areas, rotation=90)
plt.title('2016 - 2018 Cases Against Police', bbox={'facecolor': '0.8', 'pad': 5})
plt.legend(['2016', '2017', '2018'])
plt.show()
```

- <h3><b>2016 - 2018 Chargesheets Against Police Statewise Overall Comaparison [Geographical]</b></h3>

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

# Summing up chargesheets across three datasets
chargesheets = (df1.iloc[:, -1] + df2.iloc[:, -1] + df3.iloc[:, -1]).tolist()
# Creating DataFrame
df = pd.DataFrame({'Area_Name': df1.iloc[:, 0], 'Frequency': chargesheets})
# Load map file
map_file = gpd.read_file(r"E:\project-gtp-main\.mapDependencies\Indian_States.shp")
# Merge datasets
merged = map_file.set_index('st_nm').join(df.set_index('Area_Name'))
# Manual value assignment for missing areas
missing_areas = ['Andaman & Nicobar Island', 'Arunanchal Pradesh', 'Dadara & Nagar Havelli', 
                 'NCT of Delhi', 'Telangana', 'Puducherry', 'Lakshadweep', 'Sikkim', 'Uttarakhand', 'Goa']
missing_values = [14, 18, 0, 0, 0, 98, 0, 0, 9, 28]
for area, value in zip(missing_areas, missing_values):
    merged.at[area, 'Frequency'] = value
merged['Frequency'] = merged['Frequency'].astype(int)
fig, ax = plt.subplots(figsize=(12, 8))
ax.axis('off')
ax.set_title('Statewise Comparison [Chargesheets Against Police] (2001 - 2010)', 
             bbox={'facecolor': '0.8', 'pad': 5}, loc='center')
merged.plot(column='Frequency', cmap='gnuplot', linewidth=0.5, ax=ax, edgecolor='0.1', legend=True)
plt.show()
```

- <h3><b>2016 - 2018 Cases Against Police Statewise Overall Comaparison [Geographical]</b></h3>

```python
cases = []
for i in range(36):
    cases.append(df1.iloc[i, -1] + df2.iloc[i, -1] + df3.iloc[i, -1])
df = pd.DataFrame(list(zip(list(df1.iloc[:, 0]), cases)), columns = ['Area_Name', 'Frequency'])
map_file = gpd.read_file(r"E:\project-gtp-main\.mapDependencies\Indian_States.shp")
merged = map_file.set_index('st_nm').join(df.set_index('Area_Name'))
na = ['Andaman & Nicobar Island', 'Arunanchal Pradesh', 'Dadara & Nagar Havelli', 
      'NCT of Delhi', 'Telangana', 'Puducherry', 'Lakshadweep', 'Sikkim', 'Uttarakhand', 'Goa']
val = [14, 18, 0, 0, 0, 98, 0, 0, 9, 28]
j = 0
for i in na:
    merged.at[i, list(merged.columns)[1]] = val[j]
    j += 1
merged.iloc[:, -1] = merged.iloc[:, -1].astype(int)
fig, ax = plt.subplots(1, figsize=(16, 16))
ax.axis('off')
ax.set_title('2001 - 2010 Statewise Comparison [Chargesheets Against Police]', bbox={'facecolor':'0.8', 'pad':5}, loc = 'center')
fig = merged.plot(column='Frequency', cmap='gist_ncar', linewidth=0.5, ax=ax, edgecolor='0.1',legend=True)
```
