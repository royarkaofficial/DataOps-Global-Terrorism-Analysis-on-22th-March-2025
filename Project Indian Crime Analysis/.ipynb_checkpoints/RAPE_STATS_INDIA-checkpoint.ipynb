<h2><a id='division_ID2'>Data Preprocessing</a></h2>

[MENU](#division_ID0)

```python
# check the null values in the dataset
df1.isnull().sum()
```

```python
# rename the columns
df1.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
age_dict = {
    'Child Victims of Rape (Below 18 Yrs) - Below 6 Years': '<6',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above - Below 12 Years': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above - Below 16 Years': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above - Below 18 Years': '>=16-<18',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl /Child Victims': 'Total Child Victims',
    'Women Victims of Rape (Above 18 Yrs) - 18 Years & Above - Below 30 Years': '>=18-<30',
    'Women Victims of Rape (Above 18 Yrs) - 30 Years & Above - Below 45 Years': '>=30-<45',
    'Women Victims of Rape (Above 18 Yrs) - 45 Years & Above - Below 60 Years': '>=45-<60',
    'Women Victims of Rape (Above 18 Yrs) - 60 Years & Above': '>=60',
    'Women Victims of Rape (Above 18 Yrs) - Total Women/Adult Victims': 'Total Adult Victims',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above': '>=16-<18',
    'State/UT': 'Area'
}
df1.rename(columns = age_dict, inplace = True)
df1.drop([df1.columns[0]], axis = 1, inplace = True)
df1
```

```python
# check the null values in the dataset
df2.isnull().sum()
```

```python
# rename the columns
df2.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
df2.rename(columns = age_dict, inplace = True)
df2.drop([df2.columns[0]], axis = 1, inplace = True)
df2
```

```python
# check the null values in the dataset
df3.isnull().sum()
```

```python
# rename the columns
df3.drop(['Cases Reported', 'Category'], axis = 1, inplace = True)
age = {
    'Child Victims of Rape (Below 18 Yrs) - Below 6 Years': '<6',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl /Child Victims': 'Total Child Victims',
    'Child Victims of Rape (Below 18 Yrs) - 6 Years & Above ': '>=6-<12',
    'Child Victims of Rape (Below 18 Yrs) - 12 Years & Above ': '>=12-<16',
    'Child Victims of Rape (Below 18 Yrs) - 16 Years & Above ': '>=16-<18',
    'Child Victims of Rape (Below 18 Yrs) - Total Girl/Child Victims': 'Total Child Victims',
    'Women Victims of Rape (Above 18 Yrs) - 18 Years & Above - Below 30 Years': '>=18-<30',
    'Women Victims of Rape (Above 18 Yrs) - 30 Years & Above - Below 45 Years': '>=30-<45',
    'Women Victims of Rape (Above 18 Yrs) - 45 Years & Above - Below 60 Years': '>=45-<60',
    'Women Victims of Rape (Above 18 Yrs) - 60 Years & Above': '>=60',
    'Women Victims of Rape (Above 18 Yrs) - Total Women/Adult Victims': 'Total Adult Victims',
    'State/UT ': 'Area'
}
df3.rename(columns = age, inplace = True)
df3.drop([df3.columns[0]], axis = 1, inplace = True)
df3
```

<h2><a id='division_ID3'>Data Visualization</a></h2>

[MENU](#division_ID0)

- <h3><b>2016 - 2018 Overall Rape Victims Age Comparison</b></h3>

```python
import matplotlib.pyplot as plt
indices = [1, 2, 3, 4, 6, 7, 8, 9]
victims_age = []
for i in indices:
    victims_age.append(df1.iloc[:, i].sum() + df2.iloc[:, i].sum() + df3.iloc[:, i].sum())
import random
plt.figure(figsize = (9, 9))
colors = [(random.random(), random.random(), random.random()) for i in range(25)]
plt.title('2016 - 2018 Overall Rape Victims Age Comparison', bbox={'facecolor':'0.8', 'pad':10}, loc = 'center')
plt.pie(victims_age, colors = colors)
my_circle=plt.Circle((0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(['<6', '>=6-<12', '>=12-<16', '>=16-<18', '>=18-<30', '>=30-<45', '>=45-<60', '>=60'], loc = 'upper right', bbox_to_anchor = (-0.1, 1.), fontsize=25, ncol = 2)
plt.show()
```

- <h3><b>2016 - 2018 Statewise Rape Victims Class Analysis</b></h3>

```python
import numpy as np
r = np.arange(36)
w = 0.3
child = []
adult = []
total = []
for i in range(36):
    child.append(df1.iloc[i, 5] + df2.iloc[i, 5] + df3.iloc[i, 5])
    adult.append(df1.iloc[i, 10] + df2.iloc[i, 10] + df3.iloc[i, 10])
    total.append(df1.iloc[i, 11] + df2.iloc[i, 11] + df3.iloc[i, 11])
plt.figure(figsize = (16, 9))
plt.bar(r, child, w)
plt.bar(r+w, adult, w)
plt.bar(r+w*2, total, w)
plt.title('2016 - 2018 Statewse Rape Victims Age Class Analysis', bbox={'facecolor':'0.8', 'pad':10}, loc = 'center')
plt.xticks(r+w, list(df1.iloc[:, 0]), rotation = 90)
plt.legend(['Child Victims', 'Adult Victims', 'Total Victims'])
plt.show()
```

- <h3><b>2016 - 2018 Statewise Rape Victims Age Analysis</b></h3>

```python
def age_statewise(j):
    indices = [1, 2, 3, 4, 6, 7, 8, 9]
    victims_age = []
    for i in indices:
        victims_age.append(df1.iloc[j, i].sum() + df2.iloc[j, i].sum() + df3.iloc[j, i].sum())
    plt.figure(figsize = (9, 9))
    colors = [(random.random(), random.random(), random.random()) for i in range(25)]
    plt.title(df1.iloc[j, 0] + ' 2016 - 2018 Overall Rape Victims Age Comparison', bbox={'facecolor':'0.8', 'pad':10}, loc = 'center')
    plt.pie(victims_age, colors = colors)
    my_circle=plt.Circle((0,0), 0.7, color='white')
    p=plt.gcf()
    p.gca().add_artist(my_circle)
    plt.legend(['<6', '>=6-<12', '>=12-<16', '>=16-<18', '>=18-<30', '>=30-<45', '>=45-<60', '>=60'], loc = 'upper right', bbox_to_anchor = (-0.1, 1.), fontsize=25, ncol = 2)
    plt.show()
```

```python
for i in range(36):
    age_statewise(i)
```
